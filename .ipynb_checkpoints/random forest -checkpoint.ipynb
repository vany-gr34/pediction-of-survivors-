```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

class DecisionTreeNode:
    def __init__(self):
        self.feature = None
        self.threshold = None
        self.left = None
        self.right = None
        self.value = None
        
class RandomForest:
    def __init__(self, n_trees=10, max_depth=10, min_samples_split=2, criterion='gini'):
        """
        criterion : 'gini' ou 'entropy' pour déterminer la mesure d'impureté.
        """
        self.n_trees = n_trees
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split
        self.criterion = criterion  # Choix entre gini ou entropy
        self.trees = []
        self.feature_importance = {}
        
    def bootstrap_sample(self, X, y):
        n_samples = X.shape[0]
        idxs = np.random.choice(n_samples, n_samples, replace=True)
        return X[idxs], y[idxs]
    
    def gini_index(self, y):
        """Calculer l'indice de Gini pour un vecteur y."""
        proportions = np.bincount(y) / len(y)
        return 1 - np.sum([p ** 2 for p in proportions if p > 0])
    
    def entropy(self, y):
        """Calculer l'entropie pour un vecteur y."""
        proportions = np.bincount(y) / len(y)
        return -np.sum([p * np.log2(p) for p in proportions if p > 0])
    
    def split_data(self, X, y, feature, threshold):
        left_mask = X[:, feature] <= threshold
        return (X[left_mask], y[left_mask], X[~left_mask], y[~left_mask])
    
    def find_best_split(self, X, y):
        best_gain = -1
        best_feature = None
        best_threshold = None
        
        # Sélectionner le critère (gini ou entropy)
        if self.criterion == 'gini':
            impurity = self.gini_index
        elif self.criterion == 'entropy':
            impurity = self.entropy
        else:
            raise ValueError("criterion doit être 'gini' ou 'entropy'")
        
        current_impurity = impurity(y)
        
        for feature in range(X.shape[1]):
            thresholds = np.unique(X[:, feature])
            for threshold in thresholds:
                X_left, y_left, X_right, y_right = self.split_data(X, y, feature, threshold)
                
                if len(y_left) == 0 or len(y_right) == 0:
                    continue
                    
                impurity_left = impurity(y_left)
                impurity_right = impurity(y_right)
                
                n = len(y)
                n_l = len(y_left)
                n_r = len(y_right)
                
                gain = current_impurity - (n_l/n * impurity_left + n_r/n * impurity_right)
                
                if gain > best_gain:
                    best_gain = gain
                    best_feature = feature
                    best_threshold = threshold
                    
        return best_feature, best_threshold
    
    def build_tree(self, X, y, depth=0):
        n_samples = len(y)
        n_features = X.shape[1]
        
        # Critères d'arrêt
        if (depth >= self.max_depth or 
            n_samples < self.min_samples_split or 
            len(np.unique(y)) == 1):
            leaf_value = Counter(y).most_common(1)[0][0]
            node = DecisionTreeNode()
            node.value = leaf_value
            return node
        
        # Trouver la meilleure division
        feature, threshold = self.find_best_split(X, y)
        
        if feature is None:
            leaf_value = Counter(y).most_common(1)[0][0]
            node = DecisionTreeNode()
            node.value = leaf_value
            return node
        
        # Créer le nœud
        node = DecisionTreeNode()
        node.feature = feature
        node.threshold = threshold
        
        # Mettre à jour l'importance des caractéristiques
        if feature in self.feature_importance:
            self.feature_importance[feature] += 1
        else:
            self.feature_importance[feature] = 1
        
        # Division récursive
        X_left, y_left, X_right, y_right = self.split_data(X, y, feature, threshold)
        node.left = self.build_tree(X_left, y_left, depth + 1)
        node.right = self.build_tree(X_right, y_right, depth + 1)
        
        return node
    
    def fit(self, X, y):
        self.trees = []
        self.feature_importance = {}
        
        for _ in range(self.n_trees):
            X_sample, y_sample = self.bootstrap_sample(X, y)
            tree = self.build_tree(X_sample, y_sample)
            self.trees.append(tree)
            
        # Normaliser l'importance des caractéristiques
        total = sum(self.feature_importance.values())
        self.feature_importance = {k: v/total for k, v in self.feature_importance.items()}
        
    def predict_tree(self, node, X):
        if node.value is not None:
            return node.value
        
        if X[node.feature] <= node.threshold:
            return self.predict_tree(node.left, X)
        return self.predict_tree(node.right, X)
    
    def predict(self, X):
        predictions = np.zeros((X.shape[0], self.n_trees))
        
        for i, tree in enumerate(self.trees):
            for j, sample in enumerate(X):
                predictions[j, i] = self.predict_tree(tree, sample)
                
        # Vote majoritaire
        return np.array([Counter(pred).most_common(1)[0][0] for pred in predictions])

        
       

def preprocess_data(df):
    # Vérifier si 'age' et 'fare' contiennent des chaînes avant d'utiliser .str.replace
    if df['age'].dtype == 'object':
        df['age'] = df['age'].str.replace(',', '.').astype(float)
    else:
        df['age'] = df['age'].astype(float)
    
    if df['fare'].dtype == 'object':
        df['fare'] = df['fare'].str.replace(',', '.').astype(float)
    else:
        df['fare'] = df['fare'].astype(float)
    
    # Remplir les valeurs manquantes
    df['age'] = df['age'].fillna(df['age'].mean())
    df['fare'] = df['fare'].fillna(df['fare'].mean())
    df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
    
    # Encoder les variables catégorielles
    df['sex'] = df['sex'].map({'male': 0, 'female': 1})
    df['embarked'] = df['embarked'].map({'S': 0, 'C': 1, 'Q': 2})
    
    # Sélectionner les caractéristiques
    features = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']
    return df[features].values


# Chargement des données avec le bon séparateur
df = pd.read_csv('c:\\Users\\win 10\\Documents\\titanic3.csv', sep=',')

# Gérer les valeurs manquantes dans la colonne survived avant la conversion
df['survived'] = df['survived'].fillna(-1)  # On remplace les NA par -1 temporairement

# Conversion des types pour y et suppression des lignes avec survived = -1
y = df['survived'].values
mask = y != -1
y = y[mask].astype(int)

# Filtrer X également pour correspondre aux données valides de y
X = preprocess_data(df)
X = X[mask]

# Le reste du code reste identique
np.random.seed(42)
train_mask = np.random.rand(len(y)) < 0.8
X_train, X_test = X[train_mask], X[~train_mask]
y_train, y_test = y[train_mask], y[~train_mask]

# Entraîner le modèle
rf = RandomForest(n_trees=100, max_depth=10)
rf.fit(X_train, y_train)

# Faire des prédictions
y_pred = rf.predict(X_test)

# Calculer l'accuracy
accuracy = np.mean(y_pred == y_test)
print(f"Accuracy: {accuracy:.2f}")

# Visualiser l'importance des caractéristiques
features = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']
importance = [rf.feature_importance.get(i, 0) for i in range(len(features))]

plt.figure(figsize=(10, 6))
plt.bar(features, importance)
plt.title('Importance des caractéristiques')
plt.xlabel('Caractéristiques')
plt.ylabel('Importance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
rf.fit(X_train, y_train)
```

```python
def create_confusion_matrix(y_true, y_pred):
    """
    Crée une matrice de confusion 2x2 pour classification binaire
    """
    # Convertir en entiers si nécessaire
    y_true = y_true.astype(int)
    y_pred = y_pred.astype(int)
    
    # Initialiser la matrice
    matrix = np.zeros((2, 2), dtype=int)
    
    # Remplir la matrice
    for t, p in zip(y_true, y_pred):
        matrix[int(t)][int(p)] += 1
        
    return matrix

def plot_confusion_matrix(matrix):
    """
    Affiche la matrice de confusion avec une mise en forme agréable
    """
    plt.figure(figsize=(8, 6))
    plt.imshow(matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Matrice de Confusion')
    
    # Ajouter les valeurs dans les cellules
    thresh = matrix.max() / 2.
    for i in range(2):
        for j in range(2):
            plt.text(j, i, str(matrix[i, j]),
                    horizontalalignment="center",
                    color="white" if matrix[i, j] > thresh else "black")
    
    plt.xlabel('Prédictions')
    plt.ylabel('Valeurs réelles')
    plt.xticks([0, 1], ['Non survivant (0)', 'Survivant (1)'])
    plt.yticks([0, 1], ['Non survivant (0)', 'Survivant (1)'])
    
    # Ajouter les métriques
    tn, fp, fn, tp = matrix.ravel()
    precision = tp / (tp + fp) if (tp + fp) != 0 else 0
    recall = tp / (tp + fn) if (tp + fn) != 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) != 0 else 0
    print("\nMétriques détaillées:")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1-score: {f1:.2f}")
    plt.tight_layout()
    plt.show()

    # Vérifions les valeurs avant de créer la matrice
print("Valeurs uniques dans y_test:", np.unique(y_test))
print("Valeurs uniques dans y_pred:", np.unique(y_pred))

# Calculer et afficher les métriques
conf_matrix = create_confusion_matrix(y_test, y_pred)
print("\nMatrice de Confusion:")
print(conf_matrix)



# Afficher la matrice de confusion
plot_confusion_matrix(conf_matrix)

```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entraîner un modèle de forêt aléatoire
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Prédictions
y_pred = rf_model.predict(X_test)


# Calcul de l'accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)
from sklearn.metrics import precision_score, recall_score, f1_score

# Calcul du recall, de la précision et du F1 score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Affichage des résultats
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)


# Afficher la matrice de confusion de manière graphique
ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=rf_model.classes_).plot()
```
